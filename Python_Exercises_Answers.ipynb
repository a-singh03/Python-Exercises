```python
# 1
#1.1
#Eq1 : x - y + z = 4 | Eq2 : 2x + y - 3z = 0 | Eq3: x + y + z = 2

import numpy as np
from numpy import linalg as npla
a1 = np.array([[1, 2, 1], [-1, 1, 1], [1,-3,1]])
a2 = np.array([4, 0, 2])
npla.solve(a1, a2)
```

```python
#2
#2.1

car_dict = { "Car": ["ABC","DEF","GHI","JKL","MNO","PQR","STU"],
           "Drive_Type": ["Manual","Automatic","Automatic","Manual","Manual","Automatic","Manual"],
           "Car_Size": ["Medium", "Small", "Medium", "Large", "Small", "Small", "Medium"]}
car_dict
```

```python
# 2.2

car_dict.update({'Horsepower' : [180, 150, 250, 300, 220, 200, 180]})
car_dict
```

```python
#3
#3.1

import pandas as pd
car_df = pd.DataFrame(car_dict)
car_df
```

```python
# 3.2

car_dict2 = {"Mileage" : [18,20,14,12,24,22,16],
            "Engine_Displacement" : [1500,800,1000,2000,1000,800,1200]}
car_df2 = pd.DataFrame(car_dict2)
car_df = pd.concat([car_df, car_df2], axis = 1)
car_df
```

```python
# 3.3

car_df_engine_mileage = car_df[['Car', 'Engine_Displacement', 'Mileage']]
car_df_engine_mileage
```

```python
#4
#4.1

car_df_sorted_mileage = car_df.copy()
car_df_sorted_mileage.sort_values(ascending = True, by = 'Mileage')
car_df_sorted_mileage
```

```python
# 4.2

car_df_sorted_hp_mileage=car_df.copy()
car_df_sorted_hp_mileage.sort_values(['Horsepower','Mileage'],ascending=[False,True])
```

```python
# 5
#5.1

car_df_filter = car_df[(car_df['Mileage'] <= 20) & (car_df['Car_Size'] == 'Small')]
car_df_filter
```

```python
#5.2

car_df_manual = car_df[car_df['Drive_Type'] == "Manual"]
car_df_manual
```

```python
# 5.2

car_df_automatic = car_df[car_df['Drive_Type'] == "Automatic"]
car_df_automatic
```

```python
#6
#6.1

car_df_merged_inner = pd.merge(car_df_manual[['Car', 'Car_Size', 'Horsepower']], car_df_automatic[['Car', 'Car_Size', 'Mileage']], on='Car_Size', how='inner') 
car_df_merged_inner
```
