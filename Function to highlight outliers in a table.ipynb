```python
def identify_outliers(x):
    q1, q3 = np.percentile(x, [25, 75])
    iqr = q3 - q1
    lower_bound = q1 - (1 * iqr)
    upper_bound = q3 + (1 * iqr)
    l = []
    for r in x:
        if (r < lower_bound):
            l.append('background-color: {}'.format('#FFB3BA'))
        elif (r > upper_bound):
            l.append('background-color: {}'.format('#00FF00'))
        else:
            l.append('background-color: {}'.format('#33FFFF'))  
    return l
    
cars_data_raw.style.apply(identify_outliers)
```
